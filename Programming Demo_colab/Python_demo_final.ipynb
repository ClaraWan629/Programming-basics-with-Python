```python
#suggested solutions
numbers = [15, 32, 8, 45, 22, 19, 11, 3, 87, 6]

# 1. Print each number in the list
print("Printing each number in the list:")
for num in numbers:
    print(num)

# 2. Calculate the sum of all the numbers in the list
print("\nCalculating the sum of all the numbers in the list:")
total_sum = 0
for num in numbers:
    total_sum += num
print(f"The sum of all the numbers is: {total_sum}")

# 3. Find the largest and smallest numbers in the list
print("\nFinding the largest and smallest numbers in the list:")
largest = numbers[0]
smallest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num
print(f"The largest number is: {largest}")
print(f"The smallest number is: {smallest}")
```

```python
#histogram
import matplotlib.pyplot as plt
import numpy as np

# Generate random data from a normal distribution
mu, sigma = 0, 0.1  # mean and standard deviation
data = np.random.normal(mu, sigma, 1000)

# Create a histogram
plt.hist(data, bins=30, alpha=0.75, color='blue', edgecolor='black')

# Add titles and labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add a grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```

*To generate the word cloud of an uploaded file (心經.txt) in traditional Chinese (using **NotoSansCJKtc-Regular.otf** to recognise Chinese font).*

```python
!pip install wordcloud matplotlib jieba

import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to read the content of the file
def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read() #indention within open file
    return text

# Function to generate the word cloud
def generate_word_cloud(text):
    # Use jieba to segment the text
    word_list = jieba.lcut(text)
    segmented_text = ' '.join(word_list)

    # Generate the word cloud
    wordcloud = WordCloud(
        font_path='NotoSansCJKtc-Regular.otf',  # specify the path to your traditional Chinese font
        width=800,
        height=600,
        background_color='white'
    ).generate(segmented_text)

    # Display the word cloud
    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

if __name__ == '__main__':
    file_path = '心經.txt'  # specify the path to your file
    text = read_file(file_path)
    generate_word_cloud(text)
```

```python
# Python3 program to find maximum in arr[] of size n


# Function to find maximum in arr[] of size n
def largest(arr, n):

	# Initialize maximum element
	mx = arr[0]

	# Traverse array elements from second
	# and compare every element with
	# current max
	for i in range(1, n):
		if arr[i] > mx: #indention within for
			mx = arr[i] #indention within if

	return mx


# Driver Code
if __name__ == '__main__':
	arr = [20, 10, 20, 4, 100]
	n = len(arr)

	# Calculating length of an array
	Ans = largest(arr, n)

	# display max
	print("Largest in given array is", Ans)

```
